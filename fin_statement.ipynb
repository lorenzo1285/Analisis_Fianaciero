# Importación de librerías

```python
from FinRatioAnalysis import FinRatioAnalysis
import yahoo_fin.stock_info as yf
import investpy as ip
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set_style("whitegrid")
```

# Analisis de Retorno

```python
retorno = ra.Return_Ratios()
retorno
```

```python
plot = plt.figure(figsize=(10,7))
plt.title('Analisis de Retorno')
plt.xlabel('Años')
plt.ylabel('Ratio')
plt.plot(retorno['Return_Equity'], 'b.-', label='Return on Equity')
plt.plot(retorno['Return_Assets'], 'y.-', label='Return on Assets')
plt.plot(retorno['Return_CapitalEmploy'], 'r.-', label='Return on Capital Employ')
plt.legend()
plt.show()
```

# Analisis de Rentabilidad

```python
rentabilidad = ra.Margin_Ratios()
rentabilidad
```

```python
plot = plt.figure(figsize=(10,7))
plt.title('Analisis de Rentabilidad')
plt.xlabel('Años')
plt.ylabel('Ratio')
plt.plot(rentabilidad['Gross_Margin'], 'b.-', label='Gross Margin')
plt.plot(rentabilidad['Operatin_Margin'], 'y.-', label='Operating Margin')
plt.plot(rentabilidad['NetProfit_Margin'], 'r.-', label='Net Profit Margin')
plt.legend()
plt.show()
```

# Analisis de Liquidez

```python
liquidez = ra.Liquidity_Ratios()
liquidez
```

```python
sns.barplot(x='Defense_Ratio', y=liquidez.index, data=liquidez);
```

```python
plot = plt.figure(figsize=(10,7))
plt.title('Analisis de Liquidez')
plt.xlabel('Años')
plt.ylabel('Ratio')
plt.plot(liquidez['Current_Ratio'], 'b.-', label='Current Ratio')
plt.plot(liquidez['Quick_Ratio'], 'y.-', label='Quick Ratio')
plt.plot(liquidez['Cash_Ratio'], 'r.-', label='Cash Ratio')
plt.legend()
plt.show()
```

# Analisis de Endeudamiento

```python
apalancamiento = ra.Leverage_Ratios()
apalancamiento
```

```python
sns.barplot(x='DebtEquity_Ratio', y=apalancamiento.index, data=apalancamiento);
```
