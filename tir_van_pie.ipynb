```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

image.png

```python
flujo = -1000000,	150000,	250000,	400000,	500000,	200000,	300000,	500000,	120000,	540000,	340000
tasa = .03
n =  len(flujo)-1
c = flujo[0]
```

```python
TIR = round(npf.irr(flujo), 5)
TIR
```

```python
VAN = npf.npv(rate=tasa, values=flujo)
VAN.round(5)
```

```python
# Graficando el NPV en función de la tasa de descuento
def npv_irr(tasas):
    result = []
    for tasa in tasas:
        result.append(npf.npv(tasa/100,flujo))
    return result

tasas = list(range(30))
plot = plt.figure(figsize=(10,7))
plt.title("NPV y la tasa de descuento")
plt.plot(tasas, npv_irr(tasas), marker='o', label='NPV')
plt.axhline(0, color='red')
axes = plt.gca()
axes.set_ylim([-120000,500000])
plt.xticks(tasas)
plt.legend(loc='upper right')
plt.show()
```
