## Compsaracin por Ao

```python
dif_balance1 = np.divide(balance_general.iloc[:,0], balance_general.iloc[:,1])
dif_balance = np.subtract(dif_balance1, 1).to_frame()
dif_balance.rename(columns={0: 'Diferencia_20-19'}, inplace=True)
dif_balance
```

```python
sns.barplot(x='Diferencia_20-19', y=dif_balance.index, data=dif_balance);
```

```python
dif_resultados1 = np.divide(resultados.iloc[:,0], resultados.iloc[:,1])
dif_resultados = np.subtract(dif_resultados1, 1).to_frame()
dif_resultados.rename(columns={0: 'Diferencia_20-19'}, inplace=True)
dif_resultados.dropna(inplace=True)
dif_resultados
```

```python
sns.barplot(x='Diferencia_20-19', y=dif_resultados.index, data=dif_resultados);
```

```python
diff1 = np.divide(retorno.iloc[0,:], retorno.iloc[1,:])
diff = np.subtract(diff1, 1).to_frame()
diff.rename(columns={0: 'Diferencia_20-19'}, inplace=True)
```

```python
diff_rentabilidad1 = np.divide(rentabilidad.iloc[0,:] , rentabilidad.iloc[1,:])
diff_rentabilidad = np.subtract(diff_rentabilidad1, 1).to_frame()
diff_rentabilidad.rename(columns={0: 'Diferencia_20-19'}, inplace=True)
```

```python
diff_liquidez1 = np.divide(liquidez.iloc[0,:] , liquidez.iloc[1,:])
diff_liquidez = np.subtract(diff_liquidez1, 1).to_frame()
diff_liquidez.rename(columns={0: 'Diferencia_20-19'}, inplace=True)
```

```python
diff_apalancamiento1 = np.divide(apalancamiento.iloc[0,:] , apalancamiento.iloc[1,:])
diff_apalancamiento = np.subtract(diff_apalancamiento1, 1).to_frame()
diff_apalancamiento.rename(columns={0: 'Diferencia_20-19'}, inplace=True)
```

```python
diff_total = diff.append([diff_rentabilidad, diff_liquidez, diff_apalancamiento])
diff_total
```

```python
sns.barplot(x='Diferencia_20-19', y=diff_total.index, data=diff_total);
```

## Comparacin cotra Industria


```python
industry.columns
```

```python
ratio_industry = industry[['ROE (Return on equity), after tax',
                           'ROA (Return on assets)', 'Gross margin',
                           'Operating margin (Return on sales)',
                           'Profit margin', 'Current Ratio', 'Quick Ratio',
                           'Cash Ratio', 'Debt-to-equity ratio', 'Debt ratio']]
ratio_industry
```

```python
ratio_industry.rename(columns={'ROE (Return on equity), after tax': 'Return_Equity',
                               'ROA (Return on assets)':'Return_Assets',
                               'Gross margin': 'Gross_Margin',
                               'Operating margin (Return on sales)':'Operatin_Margin',
                               'Profit margin': 'NetProfit_Margin',
                               'Current Ratio':'Current_Ratio', 'Quick Ratio':'Quick_Ratio',
                               'Cash Ratio':'Cash_Ratio', 'Debt-to-equity ratio':'DebtEquity_Ratio',
                               'Debt ratio':'Debt_Ratio'}, inplace=True)
ratio_industry
```

```python
ratio_industry.columns
```

```python
pool_ibm = pd.concat([retorno, rentabilidad, liquidez, apalancamiento], axis=1)
pool_ibm = pool_ibm[['Return_Equity', 'Return_Assets', 'Gross_Margin', 'Operatin_Margin',
       'NetProfit_Margin', 'Current_Ratio', 'Quick_Ratio', 'Cash_Ratio',
       'DebtEquity_Ratio', 'Debt_Ratio']]

pool_ibm
```

```python
ibm_vs_industry = np.subtract(pool_ibm.iloc[0,:], ratio_industry.iloc[0,:]).to_frame()
ibm_vs_industry.rename(columns={0: 'IBM-Industria'}, inplace=True)
ibm_vs_industry
```

```python
sns.barplot(x='IBM-Industria', y=ibm_vs_industry.index, data=ibm_vs_industry);
```
